Import required libraries

```python
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
# get the response in the form of html
wikiurl="https://en.wikipedia.org/wiki/Economy_of_India"
table_class="infobox"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
indiatable=soup.find('table',{'class':"infobox"})
```

```python
soup.title.string
```

```python
# use the 'find_all' function to bring back all instances of the 'table'
# tag in the HTML and store in 'all_tables' variable
#'required_ table' is the table containing the required information
all_tables=soup.find_all("table")
required_table=all_tables[4]
```

```python
required_table
```

```python
#empty lists to fill the table with required columns
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
H=[]
I=[]


#for loop to append the values of each row in a column of the table
for row in required_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==9:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
        G.append(cells[6].find(text=True))
        H.append(cells[7].find(text=True))
        I.append(cells[8].find(text=True))

 #replacing "\n" for each value of every row
A = [item.replace("\n", "") for item in A]
B = [item.replace("\n", "") for item in B]
C = [item.replace("\n", "") for item in C]
D = [item.replace("\n", "") for item in C]
E = [item.replace("\n", "") for item in E]
F = [item.replace("\n", "") for item in F]
G = [item.replace("\n", "") for item in G]
H = [item.replace("\n", "") for item in H]
I = [item.replace("\n", "") for item in I]

B=[item.replace(",", "") for item in B]
C=[item.replace(",", "") for item in C]
D=[item.replace(",", "") for item in D]
E=[item.replace(",", "") for item in E]
F = [item.replace("%", "") for item in F]
G = [item.replace("%", "") for item in G]
H = [item.replace("%", "") for item in H]
I = [item.replace("%", "") for item in I]
H = [item.replace("n/a", "0") for item in H]
I = [item.replace("n/a", "0") for item in I]
```

Summarizing the data into a dataframe

```python
Eco_Ind=pd.DataFrame(A, columns=['Year'])
Eco_Ind['GDP (in Bil. US $ PPP)']=B
Eco_Ind['GDP Per Capita (in US $ PPP)']=C
Eco_Ind['GDP (in US $ nominal)']=D
Eco_Ind['GDP Per Capita(in US $ nominal)']=E
Eco_Ind['GDP Growth (real %)']=F
Eco_Ind['Inflation Rate (%)']=G
Eco_Ind['Unemployment (%)']=H
Eco_Ind['Government Debt (%)']=I
```
